# Markers

```python
m = folium.Map(location=(40,116),
        zoom_start=12,
        tiles='Stamen Terrain')
tooltip = 'Click me!!!'
folium.Marker([40,116],popup='<i>test1</i>',tooltip=tooltip,icon=folium.Icon(icon='cloud')).add_to(m)
folium.Marker([40,116.1],popup='<b>test2</b>',tooltip=tooltip,icon=folium.Icon(color='green')).add_to(m)
folium.Marker([40,116.01],popup='<b>test3</b>',tooltip=tooltip,icon=folium.Icon(color='red',icon='info-sign')).add_to(m)
m
```

# Vincent/Vega and Altair/VegaLite Markers

```python
import os

vis1 = '../Machine-Learning-with-Python/data/vis1.json'
vis2 = '../Machine-Learning-with-Python/data/vis2.json'
vis3 = '../Machine-Learning-with-Python/data/vis3.json'
```

```python

import json


m = folium.Map(
    location=[46.3014, -123.7390],
    zoom_start=7,
    tiles='Stamen Terrain'
)

popup = folium.Popup()
folium.Vega(json.load(open(vis1)),width=450, height=250).add_to(popup)
folium.Marker(
    location=[47.3489, -124.708],
    popup=popup).add_to(m)

folium.Marker(
    location=[44.639, -124.5339],
    popup=folium.Popup(max_width=450).add_child(
        folium.Vega(json.load(open(vis2)), width=450, height=250))
).add_to(m)

folium.Marker(
    location=[46.216, -124.1280],
    popup=folium.Popup(max_width=450).add_child(
        folium.Vega(json.load(open(vis3)), width=450, height=250))
).add_to(m)


m
```

# GeoJSON/TopoJSON Overlays

```python
antarctic_ice_edge = '../Machine-Learning-with-Python/data/antarctic_ice_edge.json'
antarctic_ice_shelf_topo = '../Machine-Learning-with-Python/data/antarctic_ice_shelf_topo.json'

m = folium.Map(
    location=[-59.1759, -11.6016],
    tiles='Mapbox Bright',
    zoom_start=2
)

folium.GeoJson(
    antarctic_ice_edge,
    name='geojson'
).add_to(m)

folium.TopoJson(
    open(antarctic_ice_shelf_topo),
    'objects.antarctic_ice_shelf',
    name='topojson'
).add_to(m)

folium.LayerControl().add_to(m)


m
```
